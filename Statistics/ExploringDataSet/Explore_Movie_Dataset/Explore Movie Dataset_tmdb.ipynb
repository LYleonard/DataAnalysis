**任务3.1：**对 `popularity` 最高的20名电影绘制其 `popularity` 值。

```python
popularity_top20 = movie_data.sort_values(by=['popularity'], ascending=False).head(20)
sb.barplot(data=popularity_top20, x='original_title', y='popularity', color=sb.color_palette()[0])
plt.xticks(rotation=90)
```

---
**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_data['profit'] = movie_data['revenue_adj'] - movie_data['budget_adj']
year_profit = movie_data.groupby(['release_year'])['profit'].sum()
year_profit.plot.line()
plt.ylabel('Movie Total Profit by Year')
plt.xticks(np.arange(1960,2020, 5))
```

```python
sb.countplot(data=movie_data, x='release_year', color=sb.color_palette()[0])
plt.xticks(rotation=90)
fig = plt.gcf()
fig.set_size_inches(10, 8)
```

**分析：** 

1. 由以上折线图可以看出从1960到2015年，电影净利润总体上呈上升的趋势。
2. 1980年以前电影利润整体波动较大。
3. 1995年至2000年，电影利润整体出现下滑趋势.
