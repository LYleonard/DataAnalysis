## Matplotlib - Bivariate Exploration

### Scatter plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
fuel_econ = pd.read_csv('fuel_econ.csv')
fuel_econ.head()
```

```python
plt.scatter(data=fuel_econ, x='displ', y='comb')
plt.xlabel('Displacement(l)')
plt.ylabel('Combined Fuel Eff. (npg)')
```

```python
# sb.regplot(data=fuel_econ, x='displ', y='comb', fit_reg = False) #不显示拟合直线
sb.regplot(data=fuel_econ, x='displ', y='comb')
plt.xlabel('Displacement(l)')
plt.ylabel('Combined Fuel Eff. (npg)')
```

### Transparency and Jutter for Overlap data

```python
# jutter
sb.regplot(data=fuel_econ, x='year', y='comb', x_jitter=0.3)
```

```python
# transparency and jutter
sb.regplot(data=fuel_econ, x='year', y='comb', x_jitter=0.3, 
          scatter_kws = {'alpha': 0.1})
```

```python
# transparency and jutter
sb.regplot(data=fuel_econ, x='year', y='comb', x_jitter=0.3, 
          scatter_kws = {'alpha': 0.05})
```

### Heat Maps

```python
# plt.hist2d(data=fuel_econ, x='displ', y='comb')
plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin=0.5, cmap='viridis_r')
plt.colorbar()
plt.xlabel('Displacement(l)')
plt.ylabel('Combined Fuel Eff. (npg)')
```

设置组距

```python
fuel_econ[['displ', 'comb']].describe()
```

```python
# plt.hist2d(data=fuel_econ, x='displ', y='comb')
x_bins = np.arange(0.6, 7 + 0.3, 0.3)
y_bins = np.arange(12, 58 + 3, 3)
plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin=0.5, cmap='viridis_r', bins=[x_bins, y_bins])
plt.colorbar()
plt.xlabel('Displacement(l)')
plt.ylabel('Combined Fuel Eff. (npg)')
```

### Violin plots

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered=True, categories=sedan_classes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)
```

```python
fuel_econ.head()
```

```python
base_color = sb.color_palette()[0]
# sb.violinplot(data=fuel_econ, x='VClass', y='comb', color=base_color, inner=None)
sb.violinplot(data=fuel_econ, x='VClass', y='comb')
plt.xticks(rotation = 30)
```

### Box Plots

```python
sb.boxplot(data=fuel_econ, x='VClass', y='comb')
plt.xticks(rotation = 30)
```

### Clustered Bar Charts

```python
fuel_econ['trans_type'] = fuel_econ['trans'].apply(lambda x: x.split( )[0])
```

```python
fuel_econ.head()
```

```python
ct_counts = fuel_econ.groupby(['VClass', 'trans_type']).size()
ct_counts = ct_counts.reset_index(name = 'count')
ct_counts = ct_counts.pivot(index='VClass', columns='trans_type', values='count')
```

```python
sb.heatmap(ct_counts, annot=True, fmt='d')
```

```python
sb.countplot(data=fuel_econ, x='VClass', hue='trans_type')
plt.xticks(rotation = 15)
```

### Faceted plots

```python
bins = np.arange(10, 58, 2)
g = sb.FacetGrid(data=fuel_econ, col='VClass', col_wrap=3)
g.map(plt.hist, 'comb', bins = bins)
```

### Adaptation of Univariate plots

```python
# 均值
sb.barplot(data=fuel_econ, x='VClass', y='comb')
plt.xticks(rotation = 30)
plt.ylabel('Avg. Combined Fuel Eff.')
```

```python
# 标准差
sb.barplot(data=fuel_econ, x='VClass', y='comb',ci='sd')
plt.xticks(rotation = 30)
plt.ylabel('sd. Combined Fuel Eff.')
```

```python
# 标准差
sb.pointplot(data=fuel_econ, x='VClass', y='comb',ci='sd', linestyles='')
plt.xticks(rotation = 30)
plt.ylabel('sd. Combined Fuel Eff.')
```

### Line plot

```python
bins_e = np.arange(0.6, 7+0.2, 0.2)
bins_c = bins_e[:-1] + 0.1

displ_binned = pd.cut(fuel_econ['displ'], bins_e, include_lowest=True)
comb_mean = fuel_econ['comb'].groupby(displ_binned).mean()
comb_std = fuel_econ['comb'].groupby(displ_binned).std()
```

```python
plt.errorbar(x=bins_c, y=comb_mean, yerr=comb_std)
plt.xlabel('Displacement(l)')
plt.ylabel('Combined Fuel Eff. (npg)')
```
